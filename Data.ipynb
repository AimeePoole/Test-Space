```python
import pandas as pd

data = pd.read_csv('University DataSet.csv')

df = pd.DataFrame(data)

filtered_df = df[(df['Bat Populations '] > 2500) &
                 (df['Sheep Counter'] > 50)&
                 (df['Coursework || Exams'].str.contains("Coursework", na=False))&
                 (df['Campus Accessibility'].str.contains("High", na=False))&
                 (df['Campus Astetics'].str.contains("High", na=False))]

display(filtered_df.sort_values(by=['Sheep Counter'], ascending=False))
```
